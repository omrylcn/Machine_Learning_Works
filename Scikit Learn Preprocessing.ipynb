```python
np.random.seed(1)
df = pd.DataFrame({
    'x1': np.random.normal(0, 2, 10000),
    'x2': np.random.normal(5, 3, 10000),
    'x3': np.random.normal(-5, 5, 10000)
})
```

```python
df.head()
```

```python
scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1', 'x2', 'x3'])
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,9))
ax1.set_title('Before Scaling')
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(scaled_df['x1'], ax=ax2)
sns.kdeplot(scaled_df['x2'], ax=ax2)
sns.kdeplot(scaled_df['x3'], ax=ax2)
plt.show()
```

** Learning Accuracy Change **

```python
from sklearn import svm,tree,ensemble,neighbors,naive_bayes,neural_network
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import warnings
warnings.simplefilter("ignore")
```

```python
svc=svm.SVC()
knc=neighbors.KNeighborsClassifier()
dtc=tree.DecisionTreeClassifier()
rfc=ensemble.RandomForestClassifier()
na=naive_bayes.GaussianNB()
neu=neural_network.MLPClassifier()
```

```python
# make harder case
sc_wi2=sc_wi[:,:5]
x_wine2=x_wine[:,:5]
```

```python
models={"SVC":svc,"KNeighborsClassifier":knc,"DecisionTreeClassifier":dtc,"RandomForestClassifier":rfc,
        "GaussianNB":na,"MLPClassifier":neu}
```

```python
# scaler data
#x_train,x_test,y_train,y_test = train_test_split(sc_wi2,y_wine,train_size=0.8,random_state=48)
for name,model in models.items():
    np.random.seed(12)
    scores = cross_val_score(model,sc_wi2,y_wine, cv=10)
    print(name,"Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
    
```

```python
# unsclaer data
for name,model in models.items():
    np.random.seed(12)
    scores = cross_val_score(model,x_wine2,y_wine, cv=10)
    print(name,"Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

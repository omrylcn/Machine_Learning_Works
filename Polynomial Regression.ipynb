## Polynomial Regression

```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
```

```python
m=100
x=np.linspace(-3, 3, m).reshape(m, 1)
y=2*x**6-13*x**5+26*x**4-7*x**3-28*x**2+20*x+np.random.randn(m,1)
```

```python
plt.plot(x,y)
plt.xlabel("$x_1$",fontsize=18)
plt.ylabel("$y$",rotation=0,fontsize=18)
plt.axis([-2,4,-20,20])

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_features=PolynomialFeatures(degree=26,include_bias=False)
xpol=poly_features.fit_transform(x)
print(x[:5])
xpol[:5]
```

```python
lreg=LinearRegression()
lreg.fit(xpol,y)
lreg.intercept_,lreg.coef_
```

```python
from sklearn.metrics import mean_squared_error
xn=np.linspace(-3, 3, 100).reshape(100, 1)
xnpol = poly_features.transform(xn)
yn = lreg.predict(xpol)

score=mean_squared_error(y,yn)
score=round(score,3)

plt.plot(x, y,"b.",linewidth=5)
plt.plot(x, yn, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
first_legend=plt.legend(loc="upper left", fontsize=14)
plt.text(3,-15,score)
plt.axis([-2, 4, -20, 20])

plt.show()
```
